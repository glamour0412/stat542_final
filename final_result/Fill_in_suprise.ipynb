```python
import pandas as pd
import numpy as np
from surprise import SVDpp
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import accuracy
```

```python
data_ini = pd.read_csv("/content/Feedback.csv", index_col=0)
data = data_ini.stack(dropna=False).reset_index()
data.columns = ["user_id", "item_id", "rating"]
data["item_id"] = data["item_id"].apply(lambda x: int(x[1:]) - 1)

data_no_nan = data.dropna()

reader = Reader(rating_scale=(1, 5))
data_surprise = Dataset.load_from_df(data_no_nan, reader)
trainset = data_surprise.build_full_trainset()
print("data_surprise",data_surprise)


```

```python
svdpp = SVDpp(cache_ratings=False)
svdpp.fit(trainset)


n_users = 41
n_items = 15
filled_ratings = np.empty((n_users, n_items))

for user in range(n_users):
    for item in range(n_items):
        item_name = f"V{item + 1}"
        filled_ratings[user, item] = svdpp.predict(user + 1, item).est if pd.isna(data_ini.at[user+1, f'V{item + 1}']) else data_ini.at[user+1, f'V{item + 1}']

filled_ratings_df = pd.DataFrame(filled_ratings, columns=[f"V{i + 1}" for i in range(n_items)], index=range(1, n_users + 1))
filled_ratings_df
```

```python
filled_ratings_int_df = (filled_ratings_df.round(0)).astype(int)
```
